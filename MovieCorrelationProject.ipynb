```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\Martin Padilla\Desktop\movies.csv')
```

```python
# Looking at the data
df.head()
```

```python
# Looking if any data is missing
# We got some missing data, mostly on 'Bugdet' so next im gonna drop the missing data
for column in df.columns:
    pct_missing = np.mean(df[column].isnull())
    print('{} - {}%'.format(column, round(pct_missing*100)))
```

```python
df = df.dropna()
```

```python
#Data types for our columns
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')

df.sort_values(by = ['gross'],inplace=False, ascending=False)
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
# Plotting budget vs gross income, looking for correlation between the spending and the income they generate

sns.regplot(x='budget', y='gross',data=df,scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#Looking at correlation in numerical values, we see a high correlation between gross and budget
df.corr(method='pearson')
```

```python
#Vizualizing correlation - highest ones are budget-gross and gross-votes
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation matrix for numeric only')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
